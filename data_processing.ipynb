```python
df2 = df.iloc[:, 0:6]


for i in range(100):

    if df2.loc[i, 'Q1'] == "여성":
        df2.loc[i, 'Q1'] = 1
    elif df2.loc[i, 'Q1'] == "남성":
        df2.loc[i, 'Q1'] = 2

    if df2.loc[i, 'Q3'] in ["학부 1학년", "학부 2학년", "학부 3학년", "학부 4학년"]:
        df2.loc[i, 'Q3'] = 1
    elif df2.loc[i, 'Q3'] == "대학원 과정":
        df2.loc[i, 'Q3'] = 2
    
    if df2.loc[i, 'Q4'] in ["본가에서 통학", "신촌동과 연희동 외 지역에서 자취"]:
        df2.loc[i, 'Q4'] = 1
    elif df2.loc[i, 'Q4'] == "신촌동 또는 연희동에서 자취":
        df2.loc[i, 'Q4'] = 2
    elif df2.loc[i, 'Q4'] == "교내 기숙사 거주":
        df2.loc[i, 'Q4'] = 3
    
    if df2.loc[i, 'Q5'] == "있다":
        df2.loc[i, 'Q5'] = 1
    elif df2.loc[i, 'Q5'] == "없다":
        df2.loc[i, 'Q5'] = 2
    
    if df2.loc[i, 'Q6'] == "00 ~ 33":
        df2.loc[i, 'Q6'] = 1
    elif df2.loc[i, 'Q6'] == "34 ~ 66":
        df2.loc[i, 'Q6'] = 2
    elif df2.loc[i, 'Q6'] == "67 ~ 99":
        df2.loc[i, 'Q6'] = 3

        
df2 = df2.rename(columns={'Q1':'gender', 'Q2': 'age', 'Q3': 'undergrad', 'Q4':'r_type', 'Q5': 'PM_exp', 'Q6': 'block'})

df2
```
